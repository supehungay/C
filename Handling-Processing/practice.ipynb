# PIL - The Python Imaging Library

```python
from PIL import Image
import os
```

```python
pil_im = Image.open('baby.jpg')
pil_im
```

```python
pil_im = Image.open('baby.jpg').convert('L')
pil_im
```

```python
# Create Thumbnails 
pil_im.thumbnail((128, 128))
pil_im
```

```python
# Copy and Paste Regions
box = (100, 100, 400, 400)
region = pil_im.crop(box)
region
```

```python
# Resize and Rotate
out = pil_im.resize((128, 128))
out = pil_im.rotate(45)
out
```

# Matplotlib

```python
from PIL import Image
from pylab import *
```

```python
# Plotting Images, Points, and Lines 

# read image to array
im = array(Image.open('baby.jpg'))

# plot the image
imshow(im)
```

```python
# some points
x = [100, 100, 400, 400]
y = [200, 500, 200, 500]

# plot the points widt red star-markers
plot(x, y, "r*")

# line plot connecting the first two points
plot(x[:2], y[:2])

# add title and show the plot
title("Plotting: 'baby.jpg'")
show
```

```python
# Image Contours and Histograms
from PIL import Image
from pylab import *
# read image to array
im = array(Image.open('baby.jpg').convert('L'))
# create a new figure
figure()
# don't use colors
gray()
# show contours with origin upper left corner
contour(im, origin='image')
axis('equal')
axis('off')
```

```python
figure()
hist(im.flatten(),128)
show()
```
