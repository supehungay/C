```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('./latrobe1.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
image_R = image[: , :, 0]
image_G = image[: , :, 1]
image_B = image[: , :, 2]
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.imshow(image_R, cmap='gray')
plt.subplot(132)
plt.imshow(image_G, cmap='gray')
plt.subplot(133)
plt.imshow(image_B, cmap='gray')
```

```python
def increase_brightness(image, g):
    hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    h, s, v = cv2.split(hsv)
    lim = 255 - g
    v[v > lim] = 255
    v[v <= lim] += g
    image_hsv_result = cv2.merge((h, s, v))
    image_result = cv2.cvtColor(image_hsv_result, cv2.COLOR_HSV2RGB)
    return image_result
```

```python
plt.imshow(increase_brightness(image, 50))
```

```python

def detect_edge(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
    edges = cv2.Canny(blurred_image, 50, 150)

    return edges
```

```python
plt.imshow(detect_edge(image), cmap='gray')
```
