```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Problem 1

```python
image = cv2.imread('./BrainCT.png', 0)

plt.figure(figsize=(8, 6))
plt.imshow(image, cmap='gray')
plt.title("BrainCT")
plt.axis('off')
plt.show()
```

# Problem 2

```python
plt.hist(image.ravel(), bins=64)
plt.title("Histogram")
plt.show()
```

```python
bone_image = np.where(image > 75, 1, 0).astype(np.uint8)
plt.figure(figsize=(8, 6))
plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.title("BrainCT")
plt.axis('off')
plt.subplot(122)
plt.imshow(bone_image, cmap='gray')
plt.title("Image with threshold T = 75")
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Problem 3

```python
kernel = np.ones((4, 3), np.uint8)

dilation = cv2.dilate(src=bone_image, kernel=kernel, iterations=4)
erosion = cv2.erode(dilation, kernel, iterations=4)

Imask = erosion

plt.figure(figsize=(8, 6))
plt.subplot(121)
plt.imshow(bone_image, cmap='gray')
plt.title("Image with threshold T = 75")
plt.axis('off')
plt.subplot(122)
plt.imshow(Imask, cmap='gray')
plt.title("Imask")
plt.axis('off')
plt.tight_layout()
plt.show()
```

# Problem 4

```python
kernel = np.ones((3, 3), np.uint8)
Dilate_Imask = cv2.dilate(src=Imask, kernel=kernel, iterations=1)
Subtract_Imask_1 = Dilate_Imask - Imask

plt.figure(figsize=(10, 8))

plt.subplot(131)
plt.imshow(Imask, cmap='gray')
plt.title("Imask")
plt.axis('off')

plt.subplot(132)
plt.imshow(Dilate_Imask, cmap='gray')
plt.title("Dilate Imask")
plt.axis('off')

plt.subplot(133)
plt.imshow(Subtract_Imask_1, cmap='gray') 
plt.title("Dilate_Imask - Imask")
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Problem 5

```python
kernel = np.ones((3, 3), np.uint8)
erosion_Imask = cv2.erode(src=Imask, kernel=kernel, iterations=1)
Subtract_Imask_2 = Imask - erosion_Imask

plt.figure(figsize=(10, 8))

plt.subplot(131)
plt.imshow(Imask, cmap='gray')
plt.title("Imask")
plt.axis('off')

plt.subplot(132)
plt.imshow(Imask, cmap='gray')
plt.title("Erosion Imask")
plt.axis('off')

plt.subplot(133)
plt.imshow(Subtract_Imask_2, cmap='gray') 
plt.title("Imask - Erosion_Imask")
plt.axis('off')

plt.tight_layout()
plt.show()
```
